```python
import os
import json
import requests
```

```python
M1_FILE = "data/milestone1_research_results.json"

with open(M1_FILE, "r", encoding="utf-8") as f:
    milestone1_data = json.load(f)

papers = milestone1_data["papers"]

print("‚úÖ Loaded", len(papers), "papers from Milestone 1")
```

```python
PDF_DIR = "Milestone2/pdfs"
os.makedirs(PDF_DIR, exist_ok=True)

print("üìÅ PDF directory ready:", PDF_DIR)
```

```python
def download_pdf(pdf_url, save_path):
    try:
        response = requests.get(pdf_url, timeout=20)
        if response.status_code == 200:
            with open(save_path, "wb") as f:
                f.write(response.content)
            return True
        else:
            return False
    except Exception as e:
        return False
```

```python
downloaded = 0
failed = 0

for idx, paper in enumerate(papers, start=1):
    pdf_url = paper.get("pdf_url")

    if not pdf_url:
        print(f"‚ùå No PDF link for paper {idx}")
        failed += 1
        continue

    pdf_path = os.path.join(PDF_DIR, f"paper_{idx}.pdf")

    success = download_pdf(pdf_url, pdf_path)

    if success:
        print(f"‚úÖ Downloaded: paper_{idx}.pdf")
        downloaded += 1
    else:
        print(f"‚ùå Failed to download paper {idx}")
        failed += 1

print("\nüìä Download Summary")
print("Downloaded:", downloaded)
print("Failed:", failed)
```
